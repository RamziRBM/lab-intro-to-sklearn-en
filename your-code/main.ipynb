Evaluate the data to ensure that all numeric columns are correctly detected as such by pandas. If a column is misclassified as object, coerce it to numeric.

```python
# Your code here:


# Quisk look at current dtypes:
print("Before:\n", auto.dtypes, "\n")
#Shows each column’s current type (e.g., float64, int64, object).
# Find columns parsed as 'object'(strings)

obj_columns = auto.select_dtypes(include="object").columns.tolist()
print("Object columns;", obj_columns)
# identify columns read as object; gather names of columns that pandas things are strings.
# Some of these may actually be numbers stored as strings (e.g., "130", "?").

## Keep obvious text column; convert the rest
normalize = lambda s: s.strip().lower()
text_like = {"car name", "car_name", "name"}
# a set of column names that are meant to stay as text
to_convert = [c for c in obj_columns if c.lower() not in text_like] # list comprehension
# protect truly textual columns; mark candidates for conversion
# we keep known text columns (ex. car name) as strings and plan to convers the rest to numeric

## For each column c, preview how many new NaNs coercion would create
for c in to_convert:
    converted = pd.to_numeric(auto[c], errors="coerce") # < -- 'errors', not 'error'
    newly_nan = converted.isna().sum() - auto[c].isna().sum()
    if newly_nan > 0:
        print(f"Coercing '{c}': {newly_nan} non-numeric value(s) -> NaN")

## Actually convert
for c in to_convert:
    auto[c] = pd.to_numeric(auto[c], errors="coerce")

## one liner report
report = {
    c: (pd.to_numeric(auto[c], errors="coerce").isna().sum() - auto[c].isna().sum())
    for c in to_convert
}

print(report)

```

What is the newest model year and the oldest model year?

```python
# Your code here:


# Find the model-year column (handles common name variants):
## 1. Locate the model year column
name_norm = lambda s: s.strip().lower().replace("_", " ").replace("-", " ")
    # name_norm: a tiny normalization function that:
    # strip()  -  removes leading/trailing spaces,
    # lower()  -  makes comparison case-insensitive,
    # replace("_", " ") and replace("-", " ") unify common separators.
    # This helps match columns named like: model year, Model_Year, model-year
candidates = [c for c in auto.columns if "model" in name_norm(c) and "year" in name_norm(c)]
    # scans all column names and keeps those whose normalized text contains both 'model' and 'year'
if not candidates:
    raise KeyError("Couldn't find a 'model year column.")
    # if none found, raise a clear error;
model_year_column = candidates[0]  # otherwise pick first match: model_year_column

## 2. Make sure the COLUMN is NUMERIC - in case read as an onject
auto[model_year_column] = pd.to_numeric(auto[model_year_column], errors="coerce")
    # some .csv parse numbers as string (due to say stray characters like ?)
    # pd.to_numeric(......, errors="coerce") - convirts valid numbers and turns anything non-numeric to NaN
    # using numeric dtype ensures min()/max() work correctly and ignore NaNs

## 3. Compute the newest and oldest models (max/min)
newest = int(auto[model_year_column].max())
oldest = int(auto[model_year_column].min())
    # .min() and .max() on a numeric pandas Series return the smallest and larges values, ignoring any NaNs created during coercion
    # int(...) - to print clean integers

## 4. Convert two-digit years (70-82) to full years
 # if data uses two-digit years, map 70->1970
if newest <=99 and oldest <=99:
    newest_full = 1900 + newest
    oldest_full = 1900 + oldest

print("The newest model year is:", newest_full)
print("The oldest model year is:", oldest_full)
```

We would like to generate a linear regression model that will predict mpg. To do this, first drop the `car_name` column since it does not contain any quantitative data. Next separate the dataframe to predictor and response variables. Separate those into test and training data with 80% of the data in the training set and the remainder in the test set.

Assign the predictor and response training data to `X_train` and `y_train` respectively. Similarly, assign the predictor and response test data to `X_test` and `y_test`.

```python
# Your code here:


# import pandas as pd
from sklearn.model_selection import train_test_split

### 1. Find and Drop the non-quantitative "car name" column (it is text, not numeric)
name_norm = lambda s: s.strip().lower().replace("_", " ").replace("-", " ")
car_name_columns = [c for c in auto.columns if "car" in name_norm(c) and "name" in name_norm(c)]
if car_name_columns:
    auto = auto.drop(columns=car_name_columns)
# we look through all the column names and find the one that means 'car name' (car name, car_name or Car-name)
# if we find, we drop, because it is not numeric and not useful for linear regression

### 2. Separate DataFrame to predictor (X) and response (y)----
y = auto["mpg"]                     # response
X = auto.drop(columns=["mpg"])      # predictors - all other columns except mpg(response data)

# optional: ensure predictors are numeric (good practice for linear models)
X = X.apply(pd.to_numeric, errors='coerce')

### 3. Train/Test split (80% train, 20% test) ------
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=0)
# Splits 80:20 training:test with fixed random state for reproducibility
# Assigns:
    # X_train, y_train (training predictors/response)
    # X_test, y_test (testing predictors/response)


### Quick check (optional)
print("X_train:", X_train.shape, "y_train:", y_train.shape)
print("X_test:", X_test.shape, "y_test:", y_test.shape)

```

Now we will the dataset that we processed and peform linear regression on this data to predict the mpg for each vehicle. Initialize the model in the cell below.

```python
# Your code here:


from sklearn.linear_model import LinearRegression

# initialize the model
linreg = LinearRegression()

# If plan to add scaling (often helpful), can use a pipeline instead
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

linreg = make_pipeline(StandardScaler(with_mean=True), LinearRegression())

```

Next, fit the model in the cell below.

```python
# Your code here:
linreg.fit(X_train, y_train)
y_pred = linreg.predict(X_test)
```

# Challenge 4 - Evaluate the Model

the r squared score of a model tells us how much variation is explained by the model. In a typical dataset, most observations differ from the mean. When we create a model, we are trying to generate an equation that will tell us by how much each observation will differ from the mean. Obviously, the vast majority of models are not perfect. They can only predict some of the variation from the mean but not all of it. We attribute the rest of the difference between the actual value and the mean to random error. We would like random error to explain the as little as possible of the variation. This is why the r squared score is an important metric.

In the next cell, compute the r squared score of the model. Do this by first computing the predicted values and assign them to `y_pred`.

```python
# Your code here:


from sklearn.metrics import r2_score

# 1) Use the trained model to make predictions on the *test* features
y_pred = linreg.predict(X_test)
    # feeds the test-set features X_test into fitted model
    # returns y_pred, a 1D array of predicted mpg values for test cars

# 2) Compare predictions to the true test targets with R²
r2 = r2_score(y_test, y_pred)
    # Compares predictions to the true labels.
    # R² = 1 − (SS_res / SS_tot), where:
        # SS_res = Σ(actual − predicted)² (unexplained variance)
        # SS_tot = Σ(actual − mean(actual))² (total variance)

    # Closer to 1.0 is better; 0 means “no better than using the mean”;
    # negative means worse than the mean.
print("R²:", r2)

```

#### Our next step is to evaluate the model using the test data. We would like to ensure that our model is not overfitting the data. This means that our model will not be able to generalize well outside of the training data.

In the cell below, use the model to generate the predicted values for the training data and assign them to `y_test_pred`. Compute the r squared score for the test data by comparing the oberserved `y_train` data and the predicted `y_test_pred`.

```python
# Your code here:
# To evaluate properly:
    # Predict on training features → compare to y_train (this is the training R²).
    # Predict on test features → compare to y_test (this is the test R²).

# R² (train) shows how well the model fits data it has seen.
# Predict on training set and compute R² (training performance)
y_train_pred = linreg.predict(X_train)
r2_train = r2_score(y_train, y_train_pred)

# R² (test) shows how well it generalizes to unseen data.
# Predict on test set and compute R² (generalization performance)
y_test_pred = linreg.predict(X_test)
r2_test = r2_score(y_test, y_test_pred)

# If train ≫ test, you’re likely overfitting.
# If both are low, you may be underfitting or need better features/modeling.
print("R² (train):", r2_train)
print("R² (test): ", r2_test)

# R² (train): 0.8088490656511089 < R² (test):  0.8088938602131774
```
